```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib notebook
%matplotlib inline
import math
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

#fig, ax1 = plt.subplots()
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18, 3.5))
time_plot=ax
left_plot = inset_axes(ax, height="100%", width="100%", loc='lower left', bbox_to_anchor=(0.055,0.07,0.38,0.45), bbox_transform=ax.transAxes)
right_plot = inset_axes(ax, height="100%", width="100%", loc='lower left', bbox_to_anchor=(0.60,0.07,0.38,0.45), bbox_transform=ax.transAxes)

time_limit=200

#linestyles = ['-', '--', '-', '--', '-', '--', '-', '--']
linestyles = [':', '--', '-.']

#markerstyles = ["+", "x", "o", "v", "^", "<", ">", "1"]
markerstyles = ["o", "s", "^"]
colors = ['black','green', 'red']

####################################################################
how_freq =20

lyapunov0 = np.loadtxt("Lyapunov-data2/RNN-correct.dat")


#delta_v = [0.01, 0.001,  0.0001,  0.00001]
#delta_v = [0.001,  0.0001]
delta_v = [0.001, 0.0001]

count_leg_MD=0
count_leg_RNN=0

time_plot.plot(lyapunov0[::how_freq,0], lyapunov0[::how_freq,1], label='(Verlet,0)',  linestyle=linestyles[count_leg_MD], marker= "None", markerfacecolor='none', color=colors[count_leg_MD])
count_leg_MD+=1
time_plot.plot(lyapunov0[::how_freq,0], lyapunov0[::how_freq,2], label='(RNN,0)',  linestyle="None", marker= markerstyles[count_leg_RNN], markerfacecolor='none', color=colors[count_leg_RNN])
count_leg_RNN+=1
for delta_v_ in delta_v:

    lyapunov1 = np.loadtxt("Lyapunov-data2/RNN-shift_vo={:.6f}.dat".format(delta_v_))
    
    decimal_value = "{:.0e}".format(delta_v_)[-1]

    time_plot.plot(lyapunov1[::how_freq,0], lyapunov1[::how_freq,1], label="(Verlet,$10^{-"+decimal_value+"}$)",  linestyle=linestyles[count_leg_MD], marker= "None", markerfacecolor='none', color=colors[count_leg_MD])
    time_plot.plot(lyapunov1[::how_freq,0], lyapunov1[::how_freq,1], label="(RNN,$10^{-"+decimal_value+"}$)",  linestyle="None", marker= markerstyles[count_leg_RNN], markerfacecolor='none', color=colors[count_leg_RNN])

    count_leg_MD+=1
    count_leg_RNN+=1
    #time_plot.plot(lyapunov1[::freq_array[count_leg],0], lyapunov1[::freq_array[count_leg],2], label='RNN-δp={:.0e}'.format(delta_v_),  linestyle=linestyles[count_leg], marker= markerstyles[count_leg])
    #count_leg+=1
    
#time_plot.set_xlabel('t')
time_plot.set_ylabel('Position, X(t)', fontsize=14)
time_plot.legend(loc='upper center', ncol=3)#loc='upper left'
time_plot.set_xlim(0, time_limit)
time_plot.set_ylim(0, 5.7)
#time_plot.set_title('Simulation trajectory data with LJ (m=1.0, x0=5.1) potential')

####################################################################


how_freq =100

lyapunov0 = np.loadtxt("Lyapunov-data2/correct.out")


#delta_v = [0.01, 0.001,  0.0001,  0.00001]
delta_v = [0.001, 0.0001]
#delta_v = [0.00001]

count_leg_MD=1

for delta_v_ in delta_v:
    
    lyapunov1 = np.loadtxt("Lyapunov-data2/shift_vo={:.6f}.out".format(delta_v_))

    error_data = abs(lyapunov0[:,1]-lyapunov1[:,1])
    lgnd_text=''
    decimal_value = "{:.0e}".format(delta_v_)[-1]
    #lgnd_text = ', MAE=' + "{:.2e}".format(np.average(error_data) )
    
    left_plot.plot(lyapunov0[::how_freq,0], error_data[::how_freq], label="(Verlet,$10^{-"+decimal_value+"}$)", linestyle=linestyles[count_leg_MD], marker= "None", markerfacecolor='none', color=colors[count_leg_MD])
    count_leg_MD+=1


#left_plot.set_xlabel('t')
left_plot.set_ylabel('|$\Delta x(t)$|', fontsize=14)
#left_plot.legend(loc='best')#loc='upper left'
left_plot.set_xlim(0, time_limit)
left_plot.set_ylim(10**-5, 10**1)
#left_plot.set_title('Lyapunov Instability of Velocity Verlet')
left_plot.set_yscale('log')


####################################################################


how_freq =10

lyapunov0 = np.loadtxt("Lyapunov-data2/RNN-correct.dat")


#delta_v = [0.01, 0.001,  0.0001,  0.00001]
delta_v = [0.001, 0.0001]
#delta_v = [0.00001]

count_leg_RNN=1

for delta_v_ in delta_v:

    lyapunov1 = np.loadtxt("Lyapunov-data2/RNN-shift_vo={:.6f}.dat".format(delta_v_))
    
    error_data = abs(lyapunov0[:,2]-lyapunov1[:,2])
    
    #lgnd_text = ', MAE=' + "{:.2e}".format(np.average(error_data) )
    decimal_value = "{:.0e}".format(delta_v_)[-1]
    
    right_plot.plot(lyapunov1[::how_freq,0], error_data[::how_freq], label="(RNN,$10^{-"+decimal_value+"}$)", linestyle="None", marker= markerstyles[count_leg_RNN], markerfacecolor='none', color=colors[count_leg_RNN])
    count_leg_RNN+=1


#right_plot.set_xlabel('t')
right_plot.set_ylabel('|$\Delta x(t)$|', fontsize=14)  #|X(t,0)-X(t,δp)|
#right_plot.legend(loc='best')#loc='upper left'
right_plot.set_xlim(0, time_limit)
right_plot.set_ylim(10**-5, 10**1)
#right_plot.set_title('Lyapunov Instability of RNN')
right_plot.set_yscale('log')


###########################

fig.text(0.5, 0.02, 'Time, t', ha='center', va='center', fontsize=14)
```
