## Plotting

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def spring_potential(x):
    k=1.0
    return 0.5*k*x**2

def spring_potential_force(x):
    k=1.0
    return -k*x

def lj_potential(x):
    elj=1.0
    sigma=1.0
    return 4*elj*((sigma/x)**12 - (sigma/x)**6)

def lj_potential_force(x):
    elj=1.0
    sigma=1.0
    return 48*elj/x*((sigma/x)**12 - 0.5*((sigma/x)**6))

def double_well_potential(x):
    return 0.25*x**4 - 0.5*x**2

def double_well_potential_force(x):
    return x-x**3

def rugged_potential(x):
    first_term = 1/50*( x**4 - x**3 - 16*x**2 + 4*x + 48 )
    second_term = 1/5*np.sin(30*(x+5))
    return first_term + second_term

def rugged_potential_force(x):
    first_term = 1/50*(-4*x**3 + 3*x**2 + 32*x - 4 )
    second_term = -6*np.cos(30*(x+5))
    return first_term + second_term

how_many_x=10000

xlim_left =-6
xlim_right = 6
#x=np.linspace(4.8, 5.5, 10000, endpoint=True)
x=np.linspace(xlim_left, xlim_right, how_many_x, endpoint=True)

p_spring =spring_potential(x)
f_spring =spring_potential_force(x)

xlj=np.linspace(0.001, xlim_right, how_many_x, endpoint=True)
p_lj =lj_potential(xlj)
f_lj =lj_potential_force(xlj)

p_dw =double_well_potential(x)
f_dw =double_well_potential_force(x)

p_rugged =rugged_potential(x)
f_rugged =rugged_potential_force(x)


#linestyles = ['-', '--', '-', '--', '-', '--', '-', '--']
linestyles = ['-',':', '--', '-.']

#markerstyles = ["+", "x", "o", "v", "^", "<", ">", "1"]
markerstyles = ["o", "s", "^", "p"]
colors = ['blue', 'black', 'green', 'red']
markersize=0.0

fig=plt.figure(figsize=(12, 6))
plt.plot(x, p_spring, label='Spring', color=colors[0], linestyle=linestyles[0], marker= markerstyles[0], markerfacecolor='none', markersize=markersize)
plt.plot(xlj,p_lj_right,color=colors[1], label='LJ',linestyle=linestyles[1], marker= markerstyles[1], markerfacecolor='none', markersize=markersize)
plt.plot(x, p_dw, label='Double-well', color=colors[2], linestyle=linestyles[2], marker= markerstyles[2], markerfacecolor='none', markersize=markersize)
plt.plot(x, p_rugged, label='Rugged', color=colors[3], linestyle=linestyles[3], marker= markerstyles[3], markerfacecolor='none', markersize=markersize)

plt.legend()
plt.xlabel('x')
plt.ylabel('U(x)')
#plt.yscale('log')
plt.ylim(-1.05,4)


fig=plt.figure(figsize=(12, 6))
plt.plot(x, f_spring, label='Spring', color=colors[0], linestyle=linestyles[0], marker= markerstyles[0], markerfacecolor='none', markersize=markersize)
#plt.plot(xlj, f_lj, label='LJ', color=colors[1], linestyle=linestyles[1], marker= markerstyles[1], markerfacecolor='none', markersize=markersize)
plt.plot(xlj, f_lj_right, color=colors[1], label='LJ',linestyle=linestyles[1], marker= markerstyles[1], markerfacecolor='none', markersize=markersize)
plt.plot(x, f_dw, label='Double-well', color=colors[2], linestyle=linestyles[2], marker= markerstyles[2], markerfacecolor='none', markersize=markersize)
plt.plot(x, f_rugged, label='Rugged', color=colors[3], linestyle=linestyles[3], marker= markerstyles[3], markerfacecolor='none', markersize=markersize)

#plt.legend()
plt.xlabel('x')
plt.ylabel('F(x)')
plt.ylim(-12,12)
```
